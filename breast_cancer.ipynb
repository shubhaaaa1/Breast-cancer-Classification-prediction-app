```python
import numpy as np
import pandas as pd
import sklearn.datasets as sd
from sklearn.model_selection import train_test_split,cross_val_score , GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm  import SVC
from sklearn.ensemble import RandomForestClassifier
```

Data Collection and Analysis

```python
data = sd.load_breast_cancer()
```

```python
print(data)
```

```python
bcancer_data = pd.DataFrame(data=data.data,columns=data.feature_names)
```

```python
bcancer_data['Outcome'] = data.target
```

Data Preprocessing


```python
from sklearn.preprocessing import StandardScaler
```

```python
X = bcancer_data.drop('Outcome',axis=1)
Y = bcancer_data['Outcome']
```

```python
ss = StandardScaler()
```

```python
bcancer = ss.fit_transform(X)
```

```python
print(bcancer)
```

```python
X = bcancer
```

Splitting of Data into Training Data and Test Data 

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.23,random_state=0)
```

```python
print(X.shape,X_train.shape,X_test.shape)
print(Y.shape,Y_train.shape,Y_test.shape)
```

Model training with default parameters 

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
models = [LogisticRegression(),SVC(),RandomForestClassifier()]
for model in models:
    model.fit(X_train,Y_train)
    train_pred = model.predict(X_train)
    train_data_accuracy = accuracy_score(Y_train,train_pred)
    print(f'The Accuracy score For the {model.__class__.__name__} of  Training data is : {train_data_accuracy*100}')
    test_pred = model.predict(X_test)
    test_data_accuracy = accuracy_score(Y_test,test_pred)
    print(f'The Accuracy score For the {model.__class__.__name__} of  Training data is : {test_data_accuracy*100}')
    print('-------------------------------------------------------------------------------------------')
```

Cross val Score evaluation

```python
for model in models:
    cross_score = cross_val_score(model,X,Y,cv=5)
    cross_average = sum(cross_score)/len(cross_score)
    cross_average = cross_average *100
    cross_average = round(cross_average,2)
    print(f'The Cross scores For the {model.__class__.__name__} are : {cross_score}')
    print(f'The Accuracy Percentage for the {model.__class__.__name__} is : {cross_average}')
    print('----------------------------------------------------------------------------------------------')
```

Inference : Logistic Regression is working best in most of the case. so we choose Logistic Regression

HyperParameter Tuning 

```python
LR_params = {
    'C' : [1,3,5,10,15,20]
}
model = LogisticRegression()

classifier = GridSearchCV(model,LR_params,cv=5)
results = classifier.fit(X_test,Y_test)
print(results.best_params_)
print(results.best_score_)
```

Final Model Evaluation

```python
model = LogisticRegression(C=3)
model.fit(X_train,Y_train)
train_pred = model.predict(X_train)
train_data_accuracy = accuracy_score(Y_train,train_pred)
print(f'The Accuracy score For the {model.__class__.__name__} of  Training data is : {train_data_accuracy*100}')
test_pred = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test,test_pred)
print(f'The Accuracy score For the {model.__class__.__name__} of  Training data is : {test_data_accuracy*100}')
print('-------------------------------------------------------------------------------------------')
```
